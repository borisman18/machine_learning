## Tenengrad

```python
class Init_image:
    def __init__(self, image):
        self.img = image

    def split_areas(self, l):
        """
        Разбиение изображения на тайлы.
        Для каждого из тайлов проверяется степень размытости.
        
        l - шаг сетки разбиения на тайлы.
        """
        
        w, h = self.img.size
        stride_x = round(w // l)
        stride_y = round(h // l)

        splited_image = []
        for i in range(0, w, stride_x):
            for j in range(0, h, stride_y):
                cropped = self.img.crop((i, j, i+stride_x, j+stride_y))
                cropped = np.asarray(cropped.convert('L')) # convert back to grayscale before making it a np array.
                splited_image.append((((i, j), (i+stride_x, j+stride_y)), cropped))
        return splited_image


def apply_tenengrad(img):

    # матрицы свертки для Собеля
    Sx = [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
    Sy = [[1, 2, 1], [0, 0, 0], [-1, -2, -1]]

    Gx = convolve2d(img, Sx)
    Gy = convolve2d(img, Sy)
    S = np.sqrt(np.square(Gx) + np.square(Gy))
    thr = np.max(S) * 4/5
    res = S[S > thr]
    res = np.sum(np.square(res))
    return res
    
    

def apply_tenengrad_on_grid(file_path, l, bound=1/3):
    """
    Применение метода apply_tenengrad на каждый из кусочков изображения
    
    l - шаг сетки в пикселях.
    bound - число, домножение которого на максимальный элемент определяет порог размытости 
    для данного изображения.
    """
    
    init_image = cv2.imread(file_path)
    img = Image.open(file_path)
    pict = Init_image(img)

    splitted = pict.split_areas(l)
    teng_values, blurred_areas = [], []
    
    for tile in splitted:
        teng_value = apply_tenengrad(tile[1])
        blurred_areas.append((tile[0], teng_value))
        teng_values.append(teng_value)
    threshold = max(teng_values) * bound

    for tile in blurred_areas:
        if tile[1] > threshold:
            (x1, y1), (x2, y2) = tile[0][0], tile[0][1]
            cv2.rectangle(init_image, (x1, y1), (x2, y2), (0, 127, 0), 2)
    tenengrad_image = init_image
    cv2.imwrite('tenengrad_res.png', tenengrad_image)
    return tenengrad_image


def tenengrad(file_path='blurred.png'):
    l = 20   # шаг сетки в пикселях
    tenengrad_image = apply_tenengrad_on_grid(file_path, l)
    plt.figure(figsize=(9, 7))
    plt.imshow(tenengrad_image)


tenengrad()
```

## Gray-Level Variance Normalized

```python
def apply_glvn(img):
    
    mean, stdev = cv2.meanStdDev(img)
    s = stdev[0]**2 / mean[0]
    return s[0]


def apply_glvn_on_grid(filename, l):
    init = cv2.imread(filename)
    img = Image.open(filename)
    pict = Init_image(img)
    splitted = pict.split_areas(l)
    
    glvn_values, blurred_areas = [], []
    for tile in splitted:
        glvn_curr = apply_glvn(tile[1])
        blurred_areas.append((tile[0], glvn_curr))
        glvn_values.append(glvn_curr)
    
    threshold = max(glvn_values) * 0.05
    
    for tile in blurred_areas:
        if tile[1] < threshold:
            (x1, y1), (x2, y2) = tile[0][0], tile[0][1]
            cv2.rectangle(init, (x1, y1), (x2, y2), (0, 0, 127), 2)
        else:
            (x1, y1), (x2, y2) = tile[0][0], tile[0][1]
            cv2.rectangle(init, (x1, y1), (x2, y2), (0, 255, 0), 2)
    glvn_image = init
    return glvn_image
        

def glvn(filename='blurred.png'):

    l = 10   # шаг сетки в пикселях
    glvn_image = apply_glvn_on_grid(filename, l)
    plt.figure(figsize=(9, 7))
    plt.imshow(glvn_image)


glvn()
```

# Распознавание границ

Для выбранных изображений реализовать распознавание границ геометрических фигур. Распознанные границы нанести на рисунок (другим цветом).

### Идея
1. Применяем детектор гранниц Canny (либо маску Собеля);
2. Включаем erosion для заполнения мелких пробельных участков;
3. Пробегаем по точкам контура: если векторы образуют угол близкий к 90 градусам, то считаем, что имели место
   ортогональные векторы. Если наберем 3 стороны с двумя углами близкими к 90 град., то считаем, что перед нами прямоугольник,
   который нужно дозаполнить. 

```python
image = cv2.imread('/Users/user/Desktop/ml2/example_photo.jpeg', 1)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (3, 3), 0)
canny = cv2.Canny(blurred, 100, 255, 1)


# Find contours
cnts = cv2.findContours(canny, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]


# Iterate thorugh contours and draw rectangles around contours
for c in cnts:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(image, (x, y), (x+w, y+h), (0,255,127), 1)

plt.imshow(canny)
```

```python
def angle(x1, x2, y1, y2):
    v1_theta = math.atan2(y1, x1)
    v2_theta = math.atan2(y2, x2)
    r = (v2_theta - v1_theta) * (180.0 / math.pi)
    if r < 0:
        r += 360.0
    return r


def point_to_vect(x1, x2, y1, y2):
    return x2-x1, y2-y1


class Contour:
    def __init__(self, cnt):
        self.cnt = cnt
        self.area = cv2.contourArea(cnt)
        self.approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt, True), True)

    def angles_list(self):
        angles_by_points = []
        approx = self.approx
        len_c = len(approx)
        new_cnt = np.vstack((self.approx, self.approx[:2]))
        for i in range(len_c):
            px_1 = new_cnt[i][0][0]
            py_1 = new_cnt[i][0][1]
            px_2 = new_cnt[i + 1][0][0]
            py_2 = new_cnt[i + 1][0][1]
            px_3 = new_cnt[i + 2][0][0]
            py_3 = new_cnt[i + 2][0][1]
            x1, y1 = point_to_vect(px_1, px_2, py_1, py_2)
            x2, y2 = point_to_vect(px_3, px_2, py_3, py_2)
            angles_by_points.append(([new_cnt[i + 1][0]],angle(x1, x2, y1, y2)))

        return angles_by_points

    def split_contours(self):
        contours = []
        i = 0
        cont = []
        ang = self.angles_list()
        ang.append(ang[0]) # добавляем первый элемент, чтобы замкнуть контур
        while i < len(ang):
            if ang[i][1] < 180:
                cont.append(ang[i][0])
            else:
                contours.append(cont)
                cont = []
            i += 1
        contours.append(cont)
        return contours

    def make_rectangle(self, cont):
        px_1 = cont[0][0][0]
        py_1 = cont[0][0][1]
        px_2 = cont[1][0][0]
        py_2 = cont[1][0][1]
        px_3 = cont[2][0][0]
        py_3 = cont[2][0][1]
        dx = px_3 - px_2
        dy = py_1 - py_2
        px_4 = px_1 + dx
        py_4 = py_3 + dy
        pf = np.array([[px_4, py_4]])
        three = np.vstack((cont))
        res = np.vstack((three, pf))

        return res
```
